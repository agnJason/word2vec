```python
import gensim
import os
from gensim.models import Word2Vec
os.chdir(r"C:\Users\AGN's SP\OneDrive\课件\自然语言处理\word2vec")
#读model
model = Word2Vec.load('model')
```

```python
import collections
#读取保存了的文件  若要看人名，请修改为全集_nr.txt
with open('全集.txt', encoding='utf-8') as f:
    qj = f.readlines()
c = []
for n in qj:
    c.extend(n.strip().strip('/u300').strip('/u').split(' '))
def getstopword(path):
    with open(path, encoding='utf-8') as f:
        stopword = [line.strip('\n') for line in f.readlines()]
        stopword.append(' ')
        stopword.append('\xa0')
        stopword.append('\n')
        stopword.append('\t')
    return list(set(stopword))
def getridofsw(content_fenci, stopword):
    result = [str(i) for i in content_fenci if str(i) not in stopword and not str(i).isdigit()]
    return result
#去停用词后取前500出现率高的词
stopword = getstopword(r"stopwords.txt")
c = getridofsw(c, stopword)
count = collections.Counter(c)
def inmodel(model, word):
    try:
        model[word]
        return True
    except:
        #print(word)
        return False
words = [n[0] for n in count.most_common(500) if inmodel(model, n[0])]
```

```python
from sklearn.cluster import KMeans
#聚类 初设聚成30类（k）
k = 30
clf = KMeans(n_clusters=k)
#s为每个word的类别序号
s = clf.fit_predict(model[words])
```

```python
#展示
words_ = pd.DataFrame(words)
for i in range(k):
    print(i,':',list(words_[s==i][0]))
```

```python
#找每个类别离中心最近的词 保存在center_words(k*1)
centers = {}
center_words = [0]*k
cluster_centers = clf.cluster_centers_
for n in range(k):
    centers[n]=999
    for m in range(len(words)):
        dis = sum([(cluster_centers[n][i]-model[words[m]][i])**2 for i in range(len(cluster_centers[n]))])
        #print(dis)
        if centers[n] > dis:
            center_words[n] = words[m]
            centers[n]= dis
```

可视化展示

```python
import numpy as np
import sklearn
from sklearn.manifold import TSNE  #这是一个可以展示高维数据的包
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import matplotlib
import seaborn as sns

X = np.vstack(model[words])
y = np.hstack(s)
digits_proj = TSNE(random_state=2019).fit_transform(X)  #保存高维数据处理成2维后的数据

def scatter(x, colors):
    # 保存颜色
    palette = np.array(sns.color_palette("hls", k))
    # 建立画布
    f = plt.figure(figsize=(8, 8))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40, c=palette[colors.astype(np.int)])
    plt.xlim(-1, 1)
    plt.ylim(-1, 1)
    ax.axis('off')
    ax.axis('tight')

    # 添加每个类别的标题
    txts = []
    for i in range(k):
        # 每个类别位置
        xtext, ytext = np.median(x[colors == i, :], axis=0)
        
        #以下代码尝试给每个标题多展示几个词，规则是最近词、中间词和末尾词
        try:
            w = [center_words[i],words_[s==i].iloc[int(len(words_[s==i])/2),0],words_[s==i].iloc[-1,0]]
            w = list(set(w))
            text = '，'.join(w)
        except:
            text = center_words[i]
        
        #text = center_words[i]  #要多展示词语请注释我
        txt = ax.text(xtext, ytext, text, fontsize=10)
        txt.set_path_effects([
            PathEffects.Stroke(linewidth=5, foreground="w"),
            PathEffects.Normal()])
        txts.append(txt)

    return f, ax, sc, txts

scatter(digits_proj, y)
plt.savefig('digits_tsne-generated_{}_nr.png'.format(str(k)), dpi=120)  #保存
plt.show()
```
